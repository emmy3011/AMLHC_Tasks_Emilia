Calculating the mean values of a given dataframe

Imports

```python
import pandas as pd
from sklearn import datasets
```

Definition of calcMeans with apply

```python
def calcMeans(dataframe):
    # calculating the mean of each column in the dataframe
    means = dataframe.apply(pd.Series.mean)
    return means
```

Definition of calcMeans with for-loop

```python
def calcMedians(dataframe):
    # empty list to store the medians
    medians = []
    # calculating the median of each column in the dataframe
    for column in dataframe.columns:
        median = dataframe[column].median()
        medians.append(median)

    return pd.Series(medians, index=dataframe.columns)
```

Adding the Iris-Dataset

```python

#Loading the iris dataset
iris_X, iris_y = datasets.load_iris(return_X_y=True, as_frame=True)

#Creating a DataFrame from the iris dataset
iris_df = iris_X

#Calculating the means of each column
means = calcMeans(iris_df)


```

Output

```python
print("Means of each column in the iris dataset:")
print(means)
```
